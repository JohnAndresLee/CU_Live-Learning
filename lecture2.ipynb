# Data Cleaning

```python
# Define trading time ranges as time objects
morning_start = pd.to_datetime('09:00:00').time()
morning_end = pd.to_datetime('11:30:00').time()
morning_session1_end = pd.to_datetime('10:15:00').time()  # First morning session end (before break)
morning_session2_start = pd.to_datetime('10:30:00').time()  # Second morning session start (after break)
afternoon_start = pd.to_datetime('13:30:00').time()
afternoon_end = pd.to_datetime('15:00:00').time()
night_start = pd.to_datetime('21:00:00').time()
night_end = pd.to_datetime('01:00:00').time()

# Define trading break
break_start = pd.to_datetime('10:15:00').time()
break_end = pd.to_datetime('10:30:00').time()

# filter ticks
df = df[(
    ((df['Datetime'].dt.time >= morning_start) & (df['Datetime'].dt.time <= morning_end)) |
    ((df['Datetime'].dt.time >= afternoon_start) & (df['Datetime'].dt.time <= afternoon_end)) |
    ((df['Datetime'].dt.time >= night_start) | (df['Datetime'].dt.time <= night_end))) &  
    ~((df['Datetime'].dt.time >= break_start) & (df['Datetime'].dt.time <= break_end))
    ]
```

```python
# Create conditions for the different sessions
conditions = [
    ((df['Datetime'].dt.time >= morning_start) & (df['Datetime'].dt.time <= morning_session1_end)),  
    ((df['Datetime'].dt.time >= morning_session2_start) & (df['Datetime'].dt.time <= morning_end)),  
    ((df['Datetime'].dt.time >= afternoon_start) & (df['Datetime'].dt.time <= afternoon_end)),      
    ((df['Datetime'].dt.time >= night_start) | (df['Datetime'].dt.time <= night_end))              
]

# Labels for the conditions
choices = ['Morning Session 1', 'Morning Session 2', 'Afternoon Session', 'Night Session']

# Apply conditions and create a new 'session' column
df['session'] = np.select(conditions, choices, default='Unknown')
```

# Initial Data Exploration

## Calculate vwap since open of each day

```python
multiplier = 5
df["VWAP_since_open"] = (
    df.groupby(["TradingDay"])["delta_Turnover"].cumsum()
    / df.groupby(["TradingDay"])["delta_Volume"].cumsum() / multiplier
)

change_of_vwap_since_open = df.groupby(["TradingDay"])["VWAP_since_open"].agg(
    lambda x: x.iloc[-1] / x[x.first_valid_index()]
)
```

```python
# max uptrend
print(change_of_vwap_since_open.reset_index().iloc[change_of_vwap_since_open.argmax()].TradingDay)
print(change_of_vwap_since_open.reset_index().iloc[change_of_vwap_since_open.argmax()].VWAP_since_open)
```

```python
# max downtrend
print(change_of_vwap_since_open.reset_index().iloc[change_of_vwap_since_open.argmin()].TradingDay)
print(change_of_vwap_since_open.reset_index().iloc[change_of_vwap_since_open.argmin()].VWAP_since_open)
```

```python
max_date = change_of_vwap_since_open.idxmax()
min_date = change_of_vwap_since_open.idxmin()
```

```python
def adjust_datetime_for_night_session(df):
    # If the time is over 21:00, we adjust the date to the previous day
    df['Datetime'] = df['Datetime'].apply(lambda dt: dt - pd.Timedelta(days=1) if dt.hour >= 21 else dt)
    return df['Datetime']
```

```python

def plot_vwap_since_open(df: pd.DataFrame, title: str):
    plt.figure(figsize=(10, 4))
    plt.plot(df[["Datetime", "VWAP_since_open"]].set_index("Datetime"))
    plt.title(title)
    plt.xlabel("time")
    plt.ylabel("VWAP_since_open")
    plt.xticks(rotation=10)
    plt.show()
```

```python
# max uptrend day
date = max_date
tick_df_sample = df[df.TradingDay == date]
tick_df_sample['Datetime'] = adjust_datetime_for_night_session(tick_df_sample)
# Filter out constant VWAP_since_open values
tick_df_sample = tick_df_sample[tick_df_sample["VWAP_since_open"].diff() != 0]
plot_vwap_since_open(tick_df_sample, f"VWAP_since_open increases most")
```

```python
# max downtrend day
date = min_date
tick_df_sample = df[df.TradingDay == date]
tick_df_sample['Datetime'] = adjust_datetime_for_night_session(tick_df_sample)
# Filter out constant VWAP_since_open values
tick_df_sample = tick_df_sample[tick_df_sample["VWAP_since_open"].diff() != 0]
plot_vwap_since_open(tick_df_sample, f"VWAP_since_open increases most")
```

# Generate 1-min forward return

```python
basedate = df.ActualTime.dt.date[0]
df['intdate'] = (df.ActualTime.dt.date - basedate).apply(lambda x: x.days)
df['clockLs'] = df.intdate * 24 * 3600 + df.index.hour * 3600 + df.index.minute * 60 + df.index.second + df.index.microsecond * 1e-6
```

```python
df['macro_price'] = (df['BidVolume1'] * df['AskPrice1'] + df['AskVolume1'] * df['BidPrice1']) / (df['BidVolume1'] + df['AskVolume1'])
```

```python
forward_interval = 60
df[f'forward_{forward_interval}s_idx'] = findTmValue(df.clockLs.values, tm=forward_interval, method='F')
```

```python
# 对于每一个session，找不到未来60s的收益就
df[f'mask'] = df.groupby(['TradingDay', 'session'])['clockLs'].transform(
    lambda x: findTmValue(x.values, tm=forward_interval, method='F')
)
```

```python
df[f'forward_{forward_interval}s_ret'] = df['macro_price'].iloc[df[f'forward_{forward_interval}s_idx'].values].values / df['macro_price'].values - 1
```

```python
df.loc[df['mask'] == -1][f'forward_{forward_interval}s_ret'] = np.nan
```
